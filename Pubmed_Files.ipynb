### Objective to make a dataframe with filename, date, time and size link of that file present in that website

```python
import requests
import os
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#Requesting the website
Web_URL='https://ftp.ncbi.nlm.nih.gov/pubmed/updatefiles/'
req=requests.get(Web_URL)
print(req)
```

```python
#Reading the content of page opened i.e html 
page=bs(req.content)
page
```

```python
#Creating empty lists , considerings these as dummy , i termed the variable as d_
import numpy as np
d_names=[]
d_dates=[]
```

```python
#Finding all the tags with a (anchor tags) 
atags=page.find_all('a')

#iterating over the all tags to get only the text present inside the a tag
for i in atags:
    d_names.append(i['href'])
d_names
```

We have file names ending with - .gz, .md5, .html

```python
#To get the dates of them , since they have no tags we are fetching the contents from pre tag 
pre_tag=page.pre
cont=pre_tag.contents
cont
```

```python
#Now we need only dates, so we are goin to fetch by string using re.compile
import re
d_dates=[]
d=page.find_all(string=re.compile('202')) #We got all dates into d as a list

#We are sending the elements in d by replacing and removing empty space into dummy variable date
for i in d:
    d_dates.append(i.replace('\n','').strip())
d_dates
```

```python
#Seen there are file names with 202 in the above list, so finding and removing them from the list
for i in d_dates:
    if i.endswith(('.md5','.html','.gz')):
        d_dates.remove(i)
d_dates
```

```python
Now we have two lists with Names and dates
```

```python
#Checking length of dates 
len(d_dates)
```

```python
#Checking length of names
len(d_names)
```

There is a difference of 1 . i.e there is no date for element in 0 index of names - readme file. 

```python
#We are excluding that 0 index element in file , so there will be equal elements for dataframe
d_df=pd.DataFrame({})
d_df['Names']=d_names[1:]
d_df['Dates']=d_dates
d_df
```

```python
#Now we dont require files with md5,so removing them and their respective dates. To do that step1 is to find the index of them
import numpy as np
i=np.where(d_df['Names'].str.endswith('.md5'))
i
```

```python
#We are dropping the index values from dummy data frame and saving the required into other dataframe
df=d_df.drop(d_df.index[i])
df
```

```python
#Replacing the file ending with gz to blank
df['Names']=df['Names'].str.replace('.gz','')
df['Names']
```

```python
#Splitting the element in names column , as we need ile typein other column
t=df['Names'].str.split('.',expand=True)
t
```

```python
#Creating a new column with Type as name and input of earlier variable with index 1 into it
df['Type']=t[1]
df
```

```python
df['File Name']=t[0]
```

```python
#Similarly splitting the column dates
d=df['Dates'].str.split(' ',expand=True)
d
```

```python
#Now we are making other final data frame with all the required fields  
Finaldf=pd.DataFrame({})
Finaldf['File Name']=t[0]
Finaldf['Type']=t[1]
Finaldf['Date']=d[0]
Finaldf['Time']=d[1]
Finaldf['Size']=d[4]
```

```python
Finaldf
```

```python
#Saving the final data frame file created into the csv file
Finaldf.to_csv('Pubmed-Updated.csv')
```
